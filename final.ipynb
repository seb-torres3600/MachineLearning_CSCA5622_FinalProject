### Simple Linear Regression ###

Let's try to some simple linear regression and let's maximize our R^2 and find our best predictors. 

```python
# Exclude Target
features = df.columns[:-1]
# Create dictionary with key=feature, val=R^2 value
predictors = {}

for feature in features:
    formula = f'Target ~ Q("{feature}")'
    model = smf.ols(data=df, formula=formula).fit()
    predictors[feature] = model.rsquared

# Reverse sort dict
best_predictors=sorted(predictors, key=predictors.get, reverse=True)
#best_r_squared=predictors[best_predictors[0]]
#print(f"Top predictor: \"{best_predictors[0]}\" with R^2 equal to {best_r_squared}")

# Top five predictors
top__predictors = best_predictors[:5]
for predictor in top__predictors:
    print(f"Predictor: \"{predictor}\" with R^2 = {predictors[predictor]}")
```

### Data Split for Test and Train ###

For this case, I will be using a split of 20% for my training data

```python
X = df.drop(columns="Target")
y = df["Target"]

# Split our data 
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=25, test_size=0.2)
```

### Models ###
Let's set up our models and get accuracy scores for all of them. I decided that instead of focusing on one model, attacking this data set with many, analyzing the results and playing around with a couple would be the best approach for this. 

```python
state = 25
models = {
    "LogisticRegression" : LogisticRegression(random_state=state, solver='liblinear'),
    "DecisionTreeNoMaxLeaf": DecisionTreeClassifier(max_depth=None, max_leaf_nodes=None),
    "DecisionTree2Depth": DecisionTreeClassifier(max_depth=2, max_leaf_nodes=None),
    "DecisionTree5Depth": DecisionTreeClassifier(max_depth=5, max_leaf_nodes=None),
    "DecisionTree8Leaf": DecisionTreeClassifier(max_depth=None, max_leaf_nodes=8),
    "DecisionTree3Leaf": DecisionTreeClassifier(max_depth=None, max_leaf_nodes=3),
    "DecisionTree2Depth5Leaf": DecisionTreeClassifier(max_depth=2, max_leaf_nodes=5),
    "DecisionTree5Depth3Leaf": DecisionTreeClassifier(max_depth=5, max_leaf_nodes=3),
    "AdaBoost": AdaBoostClassifier(random_state=state),
    "K-NearestNeighbor": KNeighborsClassifier(),
    "RandomForest": RandomForestClassifier(random_state=state),
    "SVC": SVC(random_state=state)
}
```

```python
model_results = []
best_model = None
best_accuracy = 0.0
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=1.0)

    full_report = {
        "name": name,
        "accuracy": accuracy,
        "precision": report['weighted avg']['precision'],
        "recall": report['weighted avg']['recall'],
        "f1-score": report['weighted avg']['f1-score']
    }
    model_results.append(full_report)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = name

results_df = pd.DataFrame(model_results)

print(results_df)

print(f"---------------------------")
print(f"Best Model: {best_model}")
print(f"Best Accuracy: {best_accuracy}")
```

From the results above, our RandomForest model definetly seemed to be our best one and it;s better in almost all categories. But let's explore some other ones that came close to it to see if we can refine anything to beat it. 

While the decision tree doesn't have the most accuracy it might be able to be it if I adjust a couple things, I'm gonna run it through a couple scenarios to see if that improves anything for my own curiosity.

```python
results = {}
best_accuracy = 0.0
best_attributes = None

max_depth = 30
max_leaf_nodes = 30
for depth in range(max_depth):
    for leaf_nodes in range(max_leaf_nodes):
        if depth == 0:
            depth = None
        if leaf_nodes == 0 or leaf_nodes == 1:
            leaf_nodes = None

        dt = DecisionTreeClassifier(max_depth=depth, max_leaf_nodes=leaf_nodes)
        dt.fit(x_train, y_train)
        y_pred = dt.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True, zero_division=1.0)
        # print(f"{depth, leaf_nodes} - {accuracy}")
        if best_accuracy < accuracy:
            best_accuracy = accuracy
            best_attributes = (depth, leaf_nodes)

print("Best attributes and accuracy")
print(f"{best_attributes} - {best_accuracy}")
        
```

While we were able to get a pretty good accuracy by picking different depths and leaf nodes. With the numbers we got from this expirement we have a very large leaf node number which will cause overfitting and poor generalization since we've allowed the model to get so complex. I think we need to discard this model and move on to something else.

How about logistic regression? Let's see if we can improve that a bit

```python
# Lets create a function to simply running our future models
def runModel(model, name):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=1.0)
    full_report = {
        "name": name,
        "accuracy": accuracy,
        "precision": report['weighted avg']['precision'],
        "recall": report['weighted avg']['recall'],
        "f1-score": report['weighted avg']['f1-score']
    }
    tmp_df = pd.DataFrame([full_report])
    print(tmp_df)

# We want better than 0.766102 for accuracy

# Lets try increasing our max iterations
runModel(LogisticRegression(random_state=state, solver='liblinear', max_iter=3000), "Improve iteration")
# Lets adjust our class weight in case we have unbalanced data
runModel(LogisticRegression(random_state=state, solver='liblinear', class_weight='balanced'), "Improve balancing")
# Lets adjust our C a bit
runModel(LogisticRegression(random_state=state, solver='liblinear', C=0.1), "Improve C")
# Lets combine them
runModel(LogisticRegression(random_state=state, solver='liblinear', C=0.1, max_iter=3000, class_weight='balanced'), "Combined")
```

While I decided to play around with several factors, at the end of it we really couldn't meet the accuracy of our random forest. 

It seems we could not get a better accuracy than before. Let's see if we can improve what our best one was

```python
# We want better than 0.7672316384180791

# Lets try to improve our Random Forest
# Increase the number of trees to potentially improve performance
runModel(RandomForestClassifier(n_estimators=200, random_state=state), "Increase N")
# Let's increase our max depth
runModel(RandomForestClassifier(max_depth=10, random_state=state), "Max Depth")
# Lets adjsut our min number of samples to split a node
runModel(RandomForestClassifier(min_samples_split=5, random_state=state), "Min Samples")
# Our data set isn't that large, let's turn off bootstraping 
runModel(RandomForestClassifier(bootstrap=False, random_state=state), "Bootsrap off")
# Let's combine our best to see if we can improve it 
runModel(RandomForestClassifier(bootstrap=False, max_depth=10, min_samples_split=5, random_state=state), "Combination")
```

By tweaking some factors we we're able to improve our accuracy a bit. The one factor that really helped our accuracy was turing bootstraping off. In this case our data set is actually not that large so when we turned bootstraping off we were able to get each tree iun the forest trained on the entired dataset instead of a bootstrapped sample. However, it's important to note that bootstrapping is a key feature of Random Forests and it helps introduce randomness and prevent overfitting. Turning it off, while it helped in this scenario, changes the fundamental nature of the algorithm and may not always lead to better results.

### Analysis/Conclusion ###

After comparing various models, the best model in terms of accuracy for this dataset is the Random Forest. But why did this model perform the best? Here are some key features that might've helped. 

1. High-dimensional data handling:

   Our dataset contains nearly 40 features, which I consider high-dimensional. Random Forests excel in such scenarios because:
   - They can effectively navigate through many features without overfitting.
   - The algorithm randomly selects a subset of features at each split, reducing the curse of dimensionality.

2. Non-linear relationships:

   Random Forests can capture complex, non-linear relationships between features and the target variable. This is crucial because real-world data often doesn't follow simple linear patterns. The ensemble of decision trees in a Random Forest can approximate these complex relationships more accurately than linear models.

3. Robustness to outliers:

   As observed earlier in our data exploration, several features in our dataset contain outliers. Random Forests are less sensitive to outliers because:
   - They use a majority voting system (for classification) or averaging (for regression) across many trees.
   - Individual trees are built on bootstrap samples, which may not include all outliers.

5. Handling mixed data types:

   Our dataset contained a mix of numerical and categorical features. Random Forests can handle both types without extensive preprocessing, unlike some other algorithms that require careful encoding of categorical variables.

6. Reduced overfitting:

   Despite having many features, Random Forests are less prone to overfitting compared to single decision trees. This is due to:
   - The random selection of features at each split.

7. No need for feature scaling:

   Unlike algorithms based on distance calculations (e.g., K-Nearest Neighbors or Support Vector Machines), Random Forests don't require feature scaling. This simplifies the preprocessing steps and reduces the risk of errors in the scaling process.
